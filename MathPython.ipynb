```python
# This is the python program I used to solve the following math problem:
#  For what values of n, [(10^n) - 1] is divisible by 13?

tens = 1
i = 0
while i < 100:
  # I wrote the program in two ways
  # The output when you run the program: Column 1 is the value of n, Columns 2 and 3 are the remainders from the 2 different ways.

  # Way 1:
  num = 10 ** i
  number = num - 1
  result = number % 13
    
  # Way 2
  result1 = (tens - 1) % 13
  print(i, result, result1)
  i = i + 1
  tens = tens * 10
```
